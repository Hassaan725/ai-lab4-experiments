# **DATA Visualization/plotting with matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,color='red')
plt.show()
```

```python
plt.plot(x,y,color='red')
plt.xlabel('sample')
plt.ylabel('Amplitude')
plt.title('sine')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.xlabel('sample')
plt.ylabel('Amplitude')
plt.title('sine')
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.scatter(x,y1,color='red')
plt.plot(x,y2)
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.title('sine')
plt.legend(['sine','cos'])

plt.show()
```

# **DATA CLEANING**

```python
import pandas as pd
df=pd.read_csv('/content/unclean_data1.csv')
df
```

```python
df.head()
```

```python
df.columns=df.columns.str.upper()
```

```python
df
```

```python
df=df.rename(columns={'DURATION':'TIME'})
df.head()
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df['TIME']=df.TIME.fillna(df.TIME.mean())
```

```python
df
```
